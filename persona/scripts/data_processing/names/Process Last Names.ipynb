```python
import glob
import csv
```

```python
last_names = []
```

```python
# Path to the directory containing the .txt files
file_path = r'last_names*.csv'

# Specify the columns you want to extract (0-based indexing)
columns_to_extract = [1, 3, 4]  # Name, Gender, Frequency


# Loop through all .txt files in the specified directory
for file_name in glob.glob(file_path):
        # Open and read each file
    with open(file_name, mode='r', newline='') as file:
        reader = csv.reader(file)
        
        # Extract and accumulate the specified columns
        for row in reader:
            last_names.append(row[0].capitalize())
```

```python
last_names = list(set(last_names))
```

```python
last_names.sort(key=str.lower)
```

```python
last_names
```

```python
# Step 5: Export the sorted results to a CSV file
output_file = r'aggregated_last_names.csv'
# Write to CSV file
with open(output_file, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Last Name"])  # Writing the header
    for last_name in last_names:
        writer.writerow([last_name])  # Each last name in a new row

print(f"CSV file saved at: {output_file}")
```
